```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import Axes
```

```python
all_types_summary_df = pd.read_excel("./output/lexical_diversity.xlsx", sheet_name="Summary")
all_types_summary_df
```

```python
news_and_blog_summary_df = pd.read_excel("./output/news_and_blog/lexical_diversity_news_and_blog.xlsx", sheet_name="Summary")
news_and_blog_summary_df
```

```python
press_release_summary_df = pd.read_excel("./output/press_release/lexical_diversity_press_release.xlsx", sheet_name="Summary")
press_release_summary_df
```

```python
social_media_summary_df = pd.read_excel("./output/social_media/lexical_diversity_social_media.xlsx", sheet_name="Summary")
social_media_summary_df
```

```python
# Create a figure with 4 subplots, consisting of 4 rows and 1 column
fig, axes = plt.subplots(
    nrows=4, 
    ncols=1, 
    figsize=(10, 30),
    squeeze=False
)

# Get the axes as a flat list so that they're easily accessible.
# These axes represent each subplot.
axes_flat: list[Axes] = axes.flatten()

fig.suptitle("Lexical diversity per year for each text type", fontsize=16)
fig.subplots_adjust(top=0.95)

def plot_data_on_ax(i: str, df: pd.DataFrame, title: str):
    years = list(df["year"])
    
    ax = axes_flat[i]

    ax.set_title(title)

    ax.set_xlabel("year")
    ax.grid(True)
    
    ax.set_xticks(list(years))
    ax.set_xticklabels([str(y) for y in years])
    ax.tick_params(axis="x", rotation=45)

    ax.set_ylabel("lexical diversity")

    # Plot years on the x axis and the mean lexical diversity on y axis,
    # with error bars representing standard deviation.
    ax.errorbar(
        x=years, 
        y=df["mean"], 
        yerr=df["std"],
    )
    ax.plot(years, df["mean"], marker="o")

# Plot each text type on its own subplot
plot_data_on_ax(0, all_types_summary_df, "All text types")
plot_data_on_ax(1, news_and_blog_summary_df, "News and blog")
plot_data_on_ax(2, press_release_summary_df, "Press release")
plot_data_on_ax(3, social_media_summary_df, "Social media")

fig.savefig("./visualization/lexical_diversity/lineplot.png")
```
