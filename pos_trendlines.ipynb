```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import Axes
```

```python
def plot_and_save_pos(df: pd.DataFrame, title: str, save_path: str):
    years = list(df["year"])

    pos_labels = list(df.columns)
    pos_labels.remove("year")

    fig, axes = plt.subplots(
        nrows=6, ncols=3, 
        figsize=(30, 50), 
        squeeze=False)

    fig.suptitle(title, fontsize=16)
    fig.subplots_adjust(top=0.95)

    axes_flat: list[Axes] = axes.flatten()

    for i, pos in enumerate(pos_labels):

        ax = axes_flat[i]
        
        ax.set_title(pos)
        ax.grid(True)

        ax.set_xlabel("year")
        ax.set_xticks(years)
        ax.set_xticklabels([str(y) for y in years])
        ax.tick_params(axis="x", rotation=45)

        ax.set_ylabel("Proportion of all tokens")

        ax.plot(years, df[pos], marker="o")
    
    plt.draw()
    fig.savefig(save_path)

    fig.show()


def load_and_plot(misinfotext_path: str, fakespeak_path: str, title: str, save_path: str):
    misinfotext_props_df = pd.read_excel(misinfotext_path, sheet_name="proportions")
    fakespeak_props_df = pd.read_excel(fakespeak_path, sheet_name="proportions")
    full_props_df = pd.concat([misinfotext_props_df, fakespeak_props_df]).fillna(0)

    plot_and_save_pos(full_props_df, title, save_path)
```

```python
load_and_plot(
    misinfotext_path="./data/MisInfoText/Analysis_output/MisInfoText_POS_frequency.xlsx",
    fakespeak_path="./data/Fakespeak-ENG/Analysis_output/Fakespeak_POS_frequency.xlsx",
    title="POS proportions per year for all text types",
    save_path="./visualization/POS/all_types.png"
)
```
