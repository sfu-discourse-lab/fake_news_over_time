```python
import os
import pandas as pd
from wordcloud import WordCloud
```

Create output folders, if don't already exist

```python
ner_output_base_dir = "./visualization/NER"
```

```python
# Body text
text_all_dir = f"{ner_output_base_dir}/text/all"
text_social_media_dir = f"{ner_output_base_dir}/text/social_media"
text_news_and_blog_dir = f"{ner_output_base_dir}/text/news_and_blog"
text_press_release_dir = f"{ner_output_base_dir}/text/press_release"

os.makedirs(text_all_dir, exist_ok=True)
os.makedirs(text_social_media_dir, exist_ok=True)
os.makedirs(text_news_and_blog_dir, exist_ok=True)
os.makedirs(text_press_release_dir, exist_ok=True)

# Headline text
headline_all_dir = f"{ner_output_base_dir}/headline/all"
headline_social_media_dir = f"{ner_output_base_dir}/headline/social_media"
headline_news_and_blog_dir = f"{ner_output_base_dir}/headline/news_and_blog"
headline_press_release_dir = f"{ner_output_base_dir}/headline/press_release"

os.makedirs(headline_all_dir, exist_ok=True)
os.makedirs(headline_social_media_dir, exist_ok=True)
os.makedirs(headline_news_and_blog_dir, exist_ok=True)
os.makedirs(headline_press_release_dir, exist_ok=True)
```

```python
def generate_wordclouds_for_years(filename: str):
    try:
        excel_file = pd.ExcelFile(filename)

        # Assuming that each sheet name corresponds to a year
        for year in excel_file.sheet_names:
            df = pd.read_excel(excel_file, sheet_name=year)
            
            frequencies = dict(zip(df["Entity"], df["Count"]))
            
            # TODO: look for another font that's open source
            wordcloud = WordCloud(
                width=800, 
                height=400)\
                .generate_from_frequencies(frequencies)

            yield year, wordcloud
    except Exception as e:
        print(e)
```

### Generate wordclouds for NER in body text

All text types

```python
for year, wordcloud in generate_wordclouds_for_years("./data/MisInfoText/Analysis_output/MisInfoText_named_entities_frequency.xlsx"):
    wordcloud.to_file(f"{text_all_dir}/{year}.png")

for year, wordcloud in generate_wordclouds_for_years("./data/Fakespeak-ENG/Analysis_output/Fakespeak_named_entities_frequency.xlsx"):
    wordcloud.to_file(f"{text_all_dir}/{year}.png")
```

```python
for year, wordcloud in generate_wordclouds_for_years("./data/MisInfoText/Analysis_output/social_media/MisInfoText_social_media_named_entities_frequency.xlsx"):
    wordcloud.to_file(f"{text_social_media_dir}/{year}.png")

for year, wordcloud in generate_wordclouds_for_years("./data/Fakespeak-ENG/Analysis_output/social_media/Fakespeak_social_media_named_entities_frequency.xlsx"):
    wordcloud.to_file(f"{text_social_media_dir}/{year}.png")
```
