```python
import os
from helpers import generate_wordclouds_for_years
```

Create output folders, if don't already exist

```python
ngrams_output_base_dir = "./visualization/ngrams/word_clouds"
```

```python
# Body text
text_all_dir = f"{ngrams_output_base_dir}/text/all"
text_social_media_dir = f"{ngrams_output_base_dir}/text/social_media"
text_news_and_blog_dir = f"{ngrams_output_base_dir}/text/news_and_blog"
text_press_release_dir = f"{ngrams_output_base_dir}/text/press_release"

os.makedirs(text_all_dir, exist_ok=True)
os.makedirs(text_social_media_dir, exist_ok=True)
os.makedirs(text_news_and_blog_dir, exist_ok=True)
os.makedirs(text_press_release_dir, exist_ok=True)

# Headline text
headline_all_dir = f"{ngrams_output_base_dir}/headline/all"
headline_social_media_dir = f"{ngrams_output_base_dir}/headline/social_media"
headline_news_and_blog_dir = f"{ngrams_output_base_dir}/headline/news_and_blog"
headline_press_release_dir = f"{ngrams_output_base_dir}/headline/press_release"

os.makedirs(headline_all_dir, exist_ok=True)
os.makedirs(headline_social_media_dir, exist_ok=True)
os.makedirs(headline_news_and_blog_dir, exist_ok=True)
os.makedirs(headline_press_release_dir, exist_ok=True)
```

### Generate wordclouds for ngrams in body text

```python
def generate_wordclouds(filename: str):
    return generate_wordclouds_for_years(filename, "ngram_text", "ngram_count")
```

All text types

```python
for year, wordcloud in generate_wordclouds("./output/ngrams.xlsx"):
    wordcloud.to_file(f"{text_all_dir}/{year}.png")
```

News and blog

```python
for year, wordcloud in generate_wordclouds("./output/news_and_blog/ngrams_news_and_blog.xlsx"):
    wordcloud.to_file(f"{text_news_and_blog_dir}/{year}.png")
```

Press release

```python
for year, wordcloud in generate_wordclouds("./output/press_release/ngrams_press_release.xlsx"):
    wordcloud.to_file(f"{text_press_release_dir}/{year}.png")
```

Social media

```python
for year, wordcloud in generate_wordclouds("./output/social_media/ngrams_social_media.xlsx"):
    wordcloud.to_file(f"{text_social_media_dir}/{year}.png")
```
