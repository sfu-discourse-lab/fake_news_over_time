```python
import pandas as pd
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
misinfotext_df = pd.read_excel("./data/MisInfoText/PolitiFact_original_modified.xlsx", sheet_name="Working")
fakespeak_df = pd.read_excel("./data/Fakespeak-ENG/Fakespeak-ENG modified.xlsx", sheet_name="Working")
```

```python
full_df = pd.concat([misinfotext_df, fakespeak_df])
full_df.head()
```

```python
full_df["doc"] = list(nlp.pipe(full_df["originalBodyText"]))
full_df.head()
```

```python
full_df["num_tokens_in_body"] = full_df["doc"].apply(lambda doc: len(doc))
full_df.head()
```

```python
pivot_df = full_df\
    .pivot_table(
        index="originalDateYear", 
        columns="originalTextType", 
        values="num_tokens_in_body", 
        aggfunc=["count", "mean"],
        margins=True,
        margins_name="TOTAL"
    )\
    .swaplevel(0, 1, axis=1)\
    .sort_index(axis=1).fillna(0)\
    .rename(columns={"count": "n texts", "mean": "avg. length"}, level=1)\
    .rename(columns={"TOTAL": "All"}, level=0)

pivot_df.index.name = None
pivot_df.columns.names = [None, None]

pivot_df
```

```python
writer = pd.ExcelWriter("./summary_stats.xlsx", engine="xlsxwriter")
pivot_df.to_excel(writer)
writer.close()
```
